In part (a) the discount factor is 0.9, which means that the importance of the cumulative future reward is 90%, so we'll take only the 90% of the value to add it to the current value plus the current reward. So as the value of gamma grows, the value grows faster too. For this reason, reaching convergence takes longer when the value of gamma is larger, in consequence, it needs more iterations to converge. In contrast if we change the discount factor to 0.5, we can see that it converges fater, but the optimal policy also changes because the importance of the cumulative future reward is lower that the initial 0.9.
