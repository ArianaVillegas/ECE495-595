The policy is the same when we use $0$ and $-1$ reward. The range of values is $[0,1]$ if we use a zero reward when the player runs out of money. The range of values is $[-1,1]$ if we use $-1$ reward when the player runs out of money. If the action set includes $0$ and we use $-1$ reward, we get the same policy.
