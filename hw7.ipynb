# Q1

### Import the Fashion MNIST dataset

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
def get_acc(model):
    model.fit(train_images, train_labels, epochs=10)
    print('\nTrain accuracy:', model.history.history['accuracy'][-1])
    test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
    print('\nTest accuracy:', test_acc)
```

## a)

```python
model_a = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])
model_a.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
get_acc(model_a)
```

## b)

```python
model_b = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10)
])
model_b.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
get_acc(model_b)
```

```python
get_acc(model_c)
```

## d)

```python
model_d = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.Flatten(input_shape=(26, 26, 32)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10)
])
model_d.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

# Q2

### Generate training data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
def gen_data(n=100, low=-10, high=10):
    # x = np.random.uniform(low,high,n)
    x = np.linspace(low, high, num=n)
    y = np.asarray([np.exp(i) for i in x])
    x = x.reshape((len(x), 1))
    y = y.reshape((len(y), 1))

    scale_x = MinMaxScaler()
    x = scale_x.fit_transform(x)
    scale_y = MinMaxScaler()
    y = scale_y.fit_transform(y)

    return x, y
```

```python
def get_accuracy(y, expected):
    diff = np.abs(y-expected)
    return np.sum(-np.floor(np.log10(diff)) - 1)
```

```python
x, y = gen_data(500)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)

plt.scatter(x,y)
plt.title('y = exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### ANN architecture

```python
model_q2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_dim=1, activation='sigmoid'),
    tf.keras.layers.Dense(2, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model_q2.compile(optimizer='adam', loss='mse')
print(model_q2.summary())
```

### Train model

```python
model_q2.fit(x_train, y_train, epochs=5000)
```

```python

train_loss = model_q2.evaluate(x_train, y_train)
print('\nTrain loss:', train_loss)
print('\nTrain acc:', get_accuracy(x_train, y_train))
```

### Test model

```python
yhat = model_q2.predict(x_test)
test_loss = model_q2.evaluate(x_test, y_test)
print('\nTest loss:', test_loss)
print('\nTest acc:', get_accuracy(x_test, y_test))
```

```python
plt.scatter(x_test-0.005, y_test, label='Actual', alpha=0.5)
plt.scatter(x_test+0.005, yhat, label='Predicted', alpha=0.5)
plt.title('y = exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
